## 1. Problem (6)

\begin{equation}\label{eq:problem1}
\begin{aligned}
\mnmz_{\bm p}\ &\bm c^\top \bm p \\
\st\ &\sum_{i=1}^np_i=1,\\
&0\le p_i,\quad\forall i=1,\dots,n,\\
&\norm{\bm p-\bm p^0}_{k}\le\eps.
\end{aligned}
\end{equation}

### 1.1. Comparison with a solver

Since the objective function of the problem above does not depend on the order of the components of $\bm p$, without loss of generality we can assume that the vector $\bm c$ is sorted. As a result of the fact that $\bm p$ has to be a probability distribution, an increase in some components of $\bm p$ has to lead to a decrease of some other components of $\bm p$ by the same margin.To minimize the objective function, one has to decrease components of $\bm p$ corresponding to the components of $\bm c$ with high values, i.e. components with large index since the vector $\bm c$ is sorted. To demonstrate this behavior, we set $\bm p^{0}$ as uniform distribution and vector $\bm c$ is generated from a normal distribution and then sorted.

````julia
n  = 20;
p0 = ones(n)/n;
c  = sort(rand(Normal(0,1), n));
ε  = 0.05;
````

````julia
p_l1_solver = Projections.minimize_linear_on_simplex_exact(p0, c, ε, 1);

p_l1   = Projections.minimize_linear_on_simplex_l1(p0, c, ε);
p_l2   = Projections.minimize_linear_on_simplex_l2(p0, c, ε);
p_lInf = Projections.minimize_linear_on_simplex_lInf(p0, c, ε);
````

To check if the `p_l1` solution given by our algorithm is correct, we compare it with the `p_l1_solver` solution given by the  [ECOS](https://github.com/JuliaOpt/ECOS.jl)  solver. To keep the presentation simple, we compare results only fo the $l_{1}$ norm. Below you can see the norm of the differences between these two solutions 

````julia
println("‖p_l1_solver - p_l1‖  = ", round(norm(p_l1_solver .- p_l1), sigdigits = sd));
````

Moreover, below we compare values of objective functions for both solutions

````julia
println("c'*p_l1_solver = ", round(c'*p_l1_solver, sigdigits = sd))
println("c'*p_l1        = ", round(c'*p_l1, sigdigits = sd))
````

and we also check if the solutions are feasible

````julia
println("Solver solution: ")
println("  ⋅ ∑p = 1:       ", round(sum(p_l1_solver), sigdigits = sd), " = ", 1)
println("  ⋅ ‖p - p0‖ ≦ ε: ", round(norm(p_l1_solver .- p0, 1), sigdigits = sd), " ≦ ", ε)
println("Our solution: ")
println("  ⋅ ∑p = 1:       ", round(sum(p_l1), sigdigits = sd), " = 1")
println("  ⋅ ‖p - p0‖ ≦ ε: ", round(norm(p_l1 .- p0,1), sigdigits = sd), " ≦ ", ε)
````

It is clear that our solution is almost identical to the solution given by the solver. The following figures show the comparison of the original uniform distribution $\bm p^{0}$ with distributions given by the projection for $l_{1}$, $l_{2}$ and $l_{\infty}$ norm.  

````julia
ylims = extrema(vcat(p0, p_l1, p_l2, p_lInf)) .+ (-0.01, +0.01)

p1 = plot(legend = :bottomleft, title = "l1", ylims = ylims) 
scatter!(1:n, p0,   label = "p0", marker = :rect)
scatter!(1:n, p_l1, label = "p",  marker = :diamond)

p2 = plot(legend = :bottomleft, title = "l2", ylims = ylims)
scatter!(1:n, p0,   label = "p0", marker = :rect)
scatter!(1:n, p_l2, label = "p",  marker = :diamond)

p3 = plot(legend = :bottomleft, title = "l_infty", ylims = ylims)
scatter!(1:n, p0,     label = "p0", marker = :rect)
scatter!(1:n, p_lInf, label = "p",  marker = :diamond)

display(plot(p1, p2, p3, layout = (1,3), size = (900, 400)))
````

### 1.2. Time comparison

In this section, we provide a simple comparison of the time complexity of the presented algorithms. The main results are presented in the figure below, where we compare the time complexity of the algorithms for the increasing length of the vector $\bm p^{0}$. The following settings of inputs and parameters were used for the experiments

- $\bm p^{0} \sim \mathit{Uniform}(0,1)$,
- $\bm c \sim \mathit{Normal}(0,1)$,
- $n \in [10^{2},10^{5}]$,
- $\varepsilon = 0.1$,

where $n$ is the length of the vector $\bm p^{0}$. For each $n$ we ran each algorithm ten times and averaged execution times. For the $l_{2}$ norm, we use three different methods of finding the root of the function.  The `bisection` and `secant` methods are general methods implemented in [Roots](https://github.com/JuliaMath/Roots.jl) package. The `newton` method is our simple implementation of the [Newton](https://en.wikipedia.org/wiki/Newton%27s_method) method for our special case.

````julia
function eval_l1(n)
    p0 = rand(Uniform(0,1), n);
    p0 /= sum(p0);
    c  = rand(Normal(0,1), n);
    ε  = 0.1;
    return Projections.minimize_linear_on_simplex_l1(p0, c, ε);
end;

function eval_l2(n, method::Symbol)
    p0 = rand(Uniform(0,1), n);
    p0 /= sum(p0);
    c  = rand(Normal(0,1), n);
    ε  = 0.1;
    return Projections.minimize_linear_on_simplex_l2(p0, c, ε, method = method);
end;

function eval_lInf(n)
    p0 = rand(Uniform(0,1), n);
    p0 /= sum(p0);
    c  = rand(Normal(0,1), n);
    ε  = 0.1;
    return Projections.minimize_linear_on_simplex_lInf(p0, c, ε);
end;

function eval_l2_philpott(n)
    p0 = rand(Uniform(0,1), n);
    p0 /= sum(p0);
    c  = rand(Normal(0,1), n);
    ε  = 0.1;
    return Projections.philpott(p0, c, ε);
end;
````

````julia
ls  = 2:0.2:5;
ns  = ceil.(Int64, 10 .^ls);

t_l1,           = timecomparison(eval_l1, ns; maxreps = 10);
t_l2_newton,    = timecomparison((n) -> eval_l2(n, :newton), ns; maxreps = 10);
t_l2_secant,    = timecomparison((n) -> eval_l2(n, :secant), ns; maxreps = 10);
t_l2_bisection, = timecomparison((n) -> eval_l2(n, :bisection), ns; maxreps = 10);
t_lInf,         = timecomparison(eval_lInf, ns; maxreps = 10);
````

````julia
p1 = plot(legend = false, title = "l1", ylabel = "t [s]") 
plot!(ns,  t_l1)

p2 = plot(legend = :topleft, title = "l2")
plot!(ns, t_l2_newton,    label = "newton")
plot!(ns, t_l2_secant,    label = "secant")
plot!(ns, t_l2_bisection, label = "bisection")

p3 = plot(legend = false, title = "l_infty")
plot!(ns, t_lInf)

display(plot(p1, p2, p3, layout = (1,3), size = (900, 400), xlabel = "n"))
````

In the figure above we see (almost) linear dependence of time on the data size $n$. For the $l_{2}$ norm we provide the comparison of our method and the algorithm from [[Philpott 2018]](https://link.springer.com/article/10.1007/s10287-018-0314-0), see figure below. Note that the computational complexity for [[Philpott 2018]](https://link.springer.com/article/10.1007/s10287-018-0314-0) is quadratic. 

````julia
ls  = 2:0.1:3;
ns  = ceil.(Int64, 10 .^ls);

t_l2_philpott, = timecomparison(eval_l2_philpott, ns; maxreps = 10);
t_l2,          = timecomparison((n) -> eval_l2(n, :newton), ns; maxreps = 10);
````

````julia
plot(xlabel = "n", ylabel = "t [s]", legend = :topleft, size = (700, 400))
plot!(ns, t_l2, label = "l2")
display(plot!(ns, t_l2_philpott, label = "l2 philpott"))
````

## 2. Problem (8) 

\begin{equation}\label{eq:problem4}
\begin{aligned}
\mnmz_{\bm p,\bm q, r}\ &\frac12\norm{\bm p - \bm p^0}^2 + \frac12\norm{\bm q - \bm q^0}^2 + \frac12(r-r^0)^2\\
\st\ &\sum_{i=1}^n p_i=\sum_{j=1}^m q_j,\\
&0\le p_i\le C_1,\quad\forall i=1,\dots,n,\\
&0\le q_j\le C_2r,\quad\forall j=1,\dots,m.
\end{aligned}
\end{equation}

### 2.1. Comparison with a solver

````julia
n  = 20;
m  = 30;
p0 = sort(rand(Normal(0,1), n));
q0 = sort(rand(Normal(0,1), m));
r0 = rand(Uniform(0,1))
C1 = 0.7;
C2 = 0.9;
````

````julia
p_solver, q_solver, r_solver = Projections.simplex_mod1_exact(p0, q0, r0, C1, C2);
p, q, r                      = Projections.simplex_mod1(p0, q0, r0, C1, C2; atol = atol);
````

To check if the `p`, `q`, `r` given by our algorithm is correct, we compare it with the `p_solver`, `q_solver`, `r_solver` solution given by the  [ECOS](https://github.com/JuliaOpt/ECOS.jl)  solver. Below you can see the norm of the differences between these two solutions 

````julia
println("‖p_solver - p‖ = ", round(norm(p_solver .- p), sigdigits = sd))
println("‖q_solver - q‖ = ", round(norm(q_solver .- q), sigdigits = sd))
println("‖r_solver - r‖ = ", round(norm(r_solver .- r), sigdigits = sd))
````

Moreover, below we compare values of objective function
\begin{equation}
L(p,q,r) = \frac12\norm{\bm p - \bm p^0}^2 + \frac12\norm{\bm q - \bm q^0}^2 + \frac12(r-r^0)^2
\end{equation}
for both solutions

````julia
L(p, p0, q, q0, r, r0) = norm(p - p0)/2 + norm(q - q0)/2 + norm(r - r0)/2
println("L(p,q,r)                      = ", round(L(p, p0, q, q0, r, r0), sigdigits = sd))
println("L(p_solver,q_solver,r_solver) = ", round(L(p_solver, p0, q_solver, q0, r_solver, r0), sigdigits = sd))
````

and we also check if the solutions are feasible

````julia
println("Solver solution: ")
println("  ⋅ ∑p = ∑q:       ", round(sum(p_solver), sigdigits = sd), " = ", round(sum(q_solver), sigdigits = sd))
println("  ⋅ min(p) ≧ 0:    ", round(minimum(p_solver), sigdigits = sd), " ≧ 0")
println("  ⋅ min(q) ≧ 0:    ", round(minimum(q_solver), sigdigits = sd), " ≧ 0")
println("  ⋅ max(p) ≦ C1:   ", round(maximum(p_solver), sigdigits = sd), " ≦ ", C1)
println("  ⋅ max(q) ≦ C2*r: ", round(maximum(q_solver), sigdigits = sd), " ≦ ", round(C2*r_solver, sigdigits = sd))
println("Our solution: ")
println("  ⋅ ∑p = ∑q:       ", round(sum(p), sigdigits = sd), " = ", round(sum(q), sigdigits = sd))
println("  ⋅ min(p) ≧ 0:    ", round(minimum(p), sigdigits = sd), " ≧ 0")
println("  ⋅ min(q) ≧ 0:    ", round(minimum(q), sigdigits = sd), " ≧ 0")
println("  ⋅ max(p) ≦ C1:   ", round(maximum(p), sigdigits = sd), " ≦ ", C1)
println("  ⋅ max(q) ≦ C2*r: ", round(maximum(q), sigdigits = sd), " ≦ ", round(C2*r, sigdigits = sd))
````

It is clear that our solution is almost identical to the solution given by the solver. The following figures show the comparison of the original inputs $\bm p^{0}$, $\bm q^{0}$, $r^{0}$ with the outputs $\bm p$, $\bm q$, $r$ given by the projection.  

````julia
p1 = plot(legend = :topleft, title = "p vs. p0")
scatter!(1:n, p0,    label = "p0",       marker = :rect)
scatter!(1:n, p,     label = "p",        marker = :diamond)

p2 = plot(legend = :topleft, title = "q vs. q0")
scatter!(1:m, q0,    label = "q0",       marker = :rect)
scatter!(1:m, q,     label = "q",        marker = :diamond)

display(plot(p1, p2, layout = (1,2), size = (900, 400)))
````

### 2.2. Time comparison

As for the problem (6), in this section we provide a simple comparison of the time complexity of the presented algorithms. The following settings of inputs and parameters were used for the experiments

- $\bm p^{0} \sim \mathit{Normal}(0,1)$,
- $\bm q^{0} \sim \mathit{Normal}(0,1)$,
- $\bm r^{0} \sim \mathit{Uniform}(0,1)$,
- $N \in [10^{2},10^{5}]$,
    - $n = \lceil 0.3\cdot N \rceil$
    - $m = \lceil 0.7\cdot N \rceil$
- $C1 = 0.7$
- $C2 = 0.9$

where $n$ is the length of the vector $\bm p^{0}$ and $m$ is the length of the vector $\bm q^{0}$. For each $N$ we ran each algorithm ten times and averaged execution times. We use two different methods of finding the root of the function: `bisection` and `secant` methods (both methods are implemented in [Roots](https://github.com/JuliaMath/Roots.jl) package)

````julia
function eval_mod1(N, method::Symbol = :newton)
    n  = ceil(Int64, 0.3*N);
    m  = N - n;
    p0 = sort(rand(Normal(0,1), n));
    q0 = sort(rand(Normal(0,1), m));
    r0 = rand(Uniform(0,1))
    C1 = 0.7;
    C2 = 0.9;
    return Projections.simplex_mod1(p0, q0, r0, C1, C2; returnstats = true, method = method, atol = atol);
end;
````

````julia
ls  = 2:0.1:5
ns  = @. ceil(Int64, 10^ls);

ts, ts_std   = timecomparison((n) -> eval_mod1(n, :secant), ns; maxreps = 10);
tb, tb_std   = timecomparison((n) -> eval_mod1(n, :bisection), ns; maxreps = 10);
````

````julia
plot(xlabel = "N", ylabel = "t [s]", legend = :topleft, size = (700, 400))
plot!(ns, ts, label = "secant")
display(plot!(ns, tb, label = "bisection"))
````

In the figure above we see (almost) linear dependence of time on the data size $N$. We can also see, that the algorithm using the `secant` method is faster than the one using the `bisection` method. On the other hand,  the algorithm using the `bisection` method has a guaranteed convergence.

## 3. Problem (9) 

\begin{equation}\label{eq:problem3}
\begin{aligned}
\mnmz_{\bm p,\bm q}\ &\frac12\norm{\bm p - \bm p^0}^2 + \frac12\norm{\bm q - \bm q^0}^2\\
\st\ &\sum_{i=1}^n p_i=\sum_{j=1}^m q_j,\\
&0\le p_i\le C_1,\quad\forall i=1,\dots,n,\\
&0\le q_j\le \frac{1}{C_2}\sum_{i=1}^n p_i,\quad\forall j=1,\dots,m.
\end{aligned}
\end{equation}

### 3.1. Comparison with a solver

````julia
n  = 20;
m  = 30;
p0 = sort(rand(Normal(0,1), n));
q0 = sort(rand(Normal(0,1), m));
C1 = 0.7;
C2 = 9;
````

````julia
p_solver, q_solver = Projections.simplex_mod2_exact(p0, q0, C1, C2);
p, q               = Projections.simplex_mod2(p0, q0, C1, C2; atol = atol);
````

To check if the `p`, `q` given by our algorithm is correct, we compare it with the `p_solver`, `q_solver` solution given by the  [ECOS](https://github.com/JuliaOpt/ECOS.jl)  solver. Below you can see the norm of the differences between these two solutions 

````julia
println("‖p_solver - p‖ = ", round(norm(p_solver .- p), sigdigits = sd))
println("‖q_solver - q‖ = ", round(norm(q_solver .- q), sigdigits = sd))
````

Moreover, below we compare values of objective function
\begin{equation}
L(p,q) = \frac12\norm{\bm p - \bm p^0}^2 + \frac12\norm{\bm q - \bm q^0}^2
\end{equation}
for both solutions

````julia
L(p, p0, q, q0) = norm(p - p0)/2 + norm(q - q0)/2
println("L(p,q)               = ", round(L(p, p0, q, q0), sigdigits = sd))
println("L(p_solver,q_solver) = ", round(L(p_solver, p0, q_solver, q0), sigdigits = sd))
````

and we also check if the solutions are feasible

````julia
println("Solver solution: ")
println("  ⋅ ∑p = ∑q:          ", round(sum(p_solver), sigdigits = sd), " = ", round(sum(q_solver), sigdigits = sd))
println("  ⋅ min(p) ≧ 0:       ", round(minimum(p_solver), sigdigits = sd), " ≧ 0")
println("  ⋅ min(q) ≧ 0:       ", round(minimum(q_solver), sigdigits = sd), " ≧ 0")
println("  ⋅ max(p) ≦ C1:      ", round(maximum(p_solver), sigdigits = sd), " ≦ ", C1)
println("  ⋅ max(q) ≦ 1/C2*∑q: ", round(maximum(q_solver), sigdigits = sd), " ≦ ", round(sum(p_solver)/C2, sigdigits = sd))
println("Our solution: ")
println("  ⋅ ∑p = ∑q:          ", round(sum(p), sigdigits = sd), " = ", round(sum(q), sigdigits = sd))
println("  ⋅ min(p) ≧ 0:       ", round(minimum(p), sigdigits = sd), " ≧ 0")
println("  ⋅ min(q) ≧ 0:       ", round(minimum(q), sigdigits = sd), " ≧ 0")
println("  ⋅ max(p) ≦ C1:      ", round(maximum(p), sigdigits = sd), " ≦ ", C1)
println("  ⋅ max(q) ≦ 1/C2*∑q: ", round(maximum(q), sigdigits = sd), " ≦ ", round(sum(p)/C2, sigdigits = sd))
````

It is clear that our solution is almost identical to the solution given by the solver.

The following figures show the comparison of the original inputs $\bm p^{0}$, $\bm q^{0}$ with the outputs $\bm p$, $\bm q$ given by the projection.  

````julia
p1 = plot(legend = :topleft, title = "p vs. p0")
scatter!(1:n, p0,    label = "p0",       marker = :rect)
scatter!(1:n, p,     label = "p",        marker = :diamond)

p2 = plot(legend = :topleft, title = "q vs. q0")
scatter!(1:m, q0,    label = "q0",       marker = :rect)
scatter!(1:m, q,     label = "q",        marker = :diamond)

display(plot(p1, p2, layout = (1,2), size = (900, 400)))
````
